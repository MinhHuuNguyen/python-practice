```python
import os
import json
from glob import glob

from tqdm import tqdm
import pandas as pd
import cv2
```

```python
train_df = pd.read_csv('data/self_driving_cars_dataset/labels_train.csv')
train_df
```

```python
train_group_df = train_df.groupby('frame').agg({
    'xmin': list, 'xmax': list, 'ymin': list, 'ymax': list, 'class_id': list
}).reset_index()
train_group_df
```

```python
for image_name in tqdm(train_group_df.frame.to_list()):
    os.system(f'cp data/self_driving_cars_dataset/images/{image_name} data/self_driving_cars_coco/images/train/')
```

```python
val_df = pd.read_csv('data/self_driving_cars_dataset/labels_val.csv')
val_df
```

```python
val_group_df = val_df.groupby('frame').agg({
    'xmin': list, 'xmax': list, 'ymin': list, 'ymax': list, 'class_id': list
}).reset_index()
val_group_df
```

```python
for image_name in tqdm(val_group_df.frame.to_list()):
    os.system(f'cp data/self_driving_cars_dataset/images/{image_name} data/self_driving_cars_coco/images/val/')
```

```python
def convert_to_coco_label_format(x1, y1, x2, y2, img_w, img_h):
    # Format: relative
    # object_index - x_center - y_center - width - height
    x = (x1 + x2) / 2 / img_w
    y = (y1 + y2) / 2 / img_h
    w = (x2 - x1) / img_w
    h = (y2 - y1) / img_h
    return x, y, w, h
```

```python
def prepare_coco_label(root, df, _type):
    for _, row in tqdm(df.iterrows()):
        img_label_list = []
        img_path = os.path.join(root, row.frame)
        img = cv2.imread(img_path)
        img_h, img_w = img.shape[0], img.shape[1]
        for xmin, xmax, ymin, ymax, class_id in zip(row.xmin, row.xmax, row.ymin, row.ymax, row.class_id):
            x, y, w, h = convert_to_coco_label_format(xmin, ymin, xmax, ymax, img_w, img_h)
            img_label_list.append(f'{class_id} {x} {y} {w} {h}')

        label_path = img_path.replace(
            'self_driving_cars_dataset/images/', f'self_driving_cars_coco/labels/{_type}/').replace(
            '.jpg', '.txt')
        with open(label_path, 'w') as file:
            file.write('\n'.join(img_label_list))
```

```python
prepare_coco_label('data/self_driving_cars_dataset/images', train_group_df, 'train')
```

```python
prepare_coco_label('data/self_driving_cars_dataset/images', val_group_df, 'val')
```
